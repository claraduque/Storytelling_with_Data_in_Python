# 1 - Exemplo 1


Todo final de ano, a equipe da sua sorveteria de encontra com seus superiores para analisar o desempenho no ano que passou e planejar o próximo ano. 

Neste ano, foi identificado que ao perder dois funcionários da equipe ao longo do ano a sorveteria perdeu muitos potenciais clientes. 

Para apresentar esses dados, um gráfico de barras foi feito com a capacidade de atendimento dos clientes mês a mês e a quantidade de clientes atendidos.

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando os dados que vamos utilizar
df = pd.read_excel("/content/drive/MyDrive/LORDE/Minicursos/Python para Data Storytelling/Exercicio 1 - Dia 2.xlsx", sheet_name= "Planilha1")
#df = pd.read_excel('Exercicio 1 - Dia 2.xlsx')
df.head()
```

```python
#Criando nosso objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(8,8),       # (largura, altura) em polegadas 
                       dpi=110)    

#Plotando nossos dados
sns.barplot(x = "Quantidade", y= "Data", hue = "Tipo", ax = ax, data = df)

#Ajustando a formatação das datas 
datas = df['Data'].dt.strftime('%m/%Y')
ax.set_yticklabels(labels=datas)

#Adicionando o Título do gráfico
ax.set_title("Capacidade de Atendimento da Sorveteria ao longo de 2019", fontsize = 18, y = 1.02)

# Adicionando a Legenda
ax.legend(ncol = 2, loc = 0)
```

Muitas vezes, principalmente quando as barras apresentam valores próximos, a leitura do gráfico de barras não é muito intuitiva e simples de se interpretar. 

Além disso, não é fácil visualizar através desse gráfico que a perda dos funcionários ocorreu em Outubro deste ano. 

Neste caso, **um gráfico de linhas** pode tornar a leitura dos dados mais simples. 

Outro ponto importante é que queremos **destacar para nossos superiores que estamos perdendo clientes ao ficarmos com a equipe reduzida**.

```python
# Criando o objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(10,6),       # (largura, altura) em polegadas 
                       dpi=110) 

# Plotando o gráfico de linhas
cores = {"Clientes":"#D85204", "Capacidade":"#004C8E"} # criando a paleta de cores para cada uma das séries presente no gráfico
sns.lineplot(x = "Data", y= "Quantidade", hue="Tipo", ax=ax, data= df, palette = cores)

# Tirando os eixos superior e direito
sns.despine()

# Titulo do  Gráfico
ax.set_title("Análise da Capacidade da Sorveteria ao longo de 2019", loc = "left", color = '#012136', fontsize = 18)

# Modificar a cor dos eixos inferior e esquerdo e dos ticks para cinza
ax.spines['bottom'].set_color('#919191')
ax.spines['left'].set_color('#919191')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

#Organizando o Eixo X
import matplotlib.dates as mdates
# Make ticks on occurrences of each month:
ax.xaxis.set_major_locator(mdates.MonthLocator())
# Get only the month to show in the x-axis:
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.text(pd.to_datetime("2018-12-25"),1620, "2019",color='#919191', fontsize = 12)

# Adicionar o label
ax.set_xlabel('',color='#919191', fontsize = 12)
ax.set_ylabel('Número de Clientes',color='#919191', fontsize = 12) 

# Formatando a legenda
ax.legend(ncol = 2)
```

Ao modificarmos o tipo do gráfico, se tornou muito mais simples perceber o problema que ocorreu a partir de Outubro deste ano. 

Alguns pontos ainda podem ser melhorados:


*   O título do gráfico- para demonstrar diretamente o que queremos destacar
*   As séries podem ser rotuladas diretamente
*   Podemos destacar no próprio gráfico que perdermos duas pessoas da equipe em outubro



```python
#Criando o objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(10,6),       # (largura, altura) em polegadas 
                       dpi=110) 

# Plotando o gráfico de linhas
cores = {"Clientes":"#D85204", "Capacidade":"#004C8E"} # a paleta de cores para cada uma das séries presentes no gráfico
sns.lineplot(x = "Data", y= "Quantidade", hue="Tipo", ax=ax, data= df, palette = cores)

# Tirando os eixos superior e direito
sns.despine()

# Titulo do  Gráfico
ax.set_title("Equipe Desfalcada - Perda de Clientes", loc = "left", color = '#012136', fontsize= 16, y = 1.1)

# Modificar a cor dos eixos inferior e esquerdo  e os ticks para cinza
ax.spines['bottom'].set_color('#919191')
ax.spines['left'].set_color('#919191')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

#Organizando o Eixo X
import matplotlib.dates as mdates
# Make ticks on occurrences of each month:
ax.xaxis.set_major_locator(mdates.MonthLocator())
# Get only the month to show in the x-axis:
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.text(pd.to_datetime("2018-12-25"),1620, "2019",color='#919191', fontsize = 12)

# Adicionar o label
ax.set_ylabel('Número de Clientes',color='#919191', fontsize = 12)
ax.set_xlabel('',color='#919191', fontsize = 12) 

# Adicionando uma linha ao mês que ocorre a perda de funcionários para enfatizar
ax.axvline(x=pd.to_datetime("2019-10"), linestyle = "--", color ="#919191" )   
# da mesma forma que a linha vertical, existe a linha horizontal com o comando axhline

# Retirando a legenda do gráfico e rotulando as séries nelas próprias 
ax.legend().set_visible(False)
ax.text(pd.to_datetime("2019-12"), 2000, "Possíveis Clientes", fontsize = 12, color = "#D85204")
ax.text(pd.to_datetime("2019-12"), 1710, "Capacidade de \nAtendimento", fontsize = 12, color = "#004C8E")

#Adicionando a descrição da linha de destaque que colocamos
ax.text(pd.to_datetime("2019-10"), 2150, "Saída de dois funcionários", fontsize = 12, color = "#919191")

#Adicionando o texto explicativo para nosso gráfico
ax.text(pd.to_datetime("2018-12-15"), 2155, "Em Outubro deste ano, dois funcionários deixaram nossa \nloja comprometendo os dois turnos de atendimento", fontsize = 10, color = "#919191")
```

# 2 - Exemplo 2

Neste caso, foi encomendada uma pesquisa que registrava o percentual de satisfação dos clientes em diversas categorias em relação à nossa sorveteria e a sorverteria concorrente. 

A satisfação do cliente poderia variar entre 0% - totalmente insatisfeito- a 100%- totalmente satisfeito. 

Dentre as categorias participantes da pesquisa existiam perguntas relacionadas aos produtos como: sabor, textura, aroma, aparência. 
A disponibilidade dos produção oferecidos, como: diversidade de sabores e complementos. 
Além de perguntas relacionadas ao atendimento e ambiente da loja. 

**Nosso objetivo é elaborar uma apresentação/ conjunto de gráficos que destaque os resultados da pesquisa. Ressaltando nossos pontos fortes e os pontos que precisamos melhorar para nos posicionarmos em relação a concorrência.**


```python
#Importando os dados que vamos utilizar
df_2 = pd.read_excel("/content/drive/MyDrive/LORDE/Minicursos/Python para Data Storytelling/Exercicio 1 - Dia 2.xlsx", sheet_name= "Planilha2")
#df_2 = pd.read_excel("Exercicio 1 - Dia 2.xlsx", sheet_name= "Planilha2")
df_2.head()
```

Durante a elaboração da apresentação, seu colega de equipe montou um gráfico de dispersão para comparar a média de satisfação entre todas as categorias da nossa sorveteria e da concorrência. 

Seria esse tipo de gráfico o melhor para visualizar esses dados ?

```python
#Criando o objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110) 

# Plotando as duas séries da pesquisa de satisfação   
sns.scatterplot(x = "Categoria", y = "Nossa Sorveteria ", data = df_2, ax = ax , label = "Nossa Loja" )
sns.scatterplot(x = "Categoria", y = "Concorrência ", data = df_2, ax = ax , label = "Concorrência" )

# Organizando os ticks do eixo - x e o rotulo do eixo-y
ax.tick_params(axis='x', rotation=90)
ax.set_ylabel('Percentual de Satisfação(%)') 

# Titulo do gráfico
ax.set_title("Pesquisa de Satisfação da nossa loja frente à concorrência", fontsize = 16)

#Adicionando a legenda
ax.legend(loc = 4)
```

Pontos levantados em relação ao scatter plot:


*   É difícil a leitura das categorias no eixo x
*   A comparação entre os pontos não é tão intuitiva 
*   A unidade do eixo-y não está em percentual
*   É dificil concluir rapidamente quais são os pontos fortes da nossa loja


Decidimos então, dispor os dados em um gráfico de barras. 



```python
# Criando o objeto 
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110)    

# Transformando as duas séries em gráficos de barra horizontal
sns.barplot(y = "Categoria", x = "Nossa Sorveteria ", data = df_2, ax = ax , label = "Nossa Loja" )
sns.barplot(y = "Categoria", x = "Concorrência ", data = df_2, ax = ax , label = "Concorrência" )

# Organizando os rótulos dos eixos 
ax.set_ylabel('Categorias') 
ax.set_xlabel('Percentual de Satisfação') 

# Título do Gráfico
ax.set_title("Pesquisa de Satisfação da nossa loja frente à concorrência", fontsize = 18)
```

A leitura das categorias no eixo y já se tornou muito mais simples. 

Ainda é preciso consertar a unidade do agora eixo-x. 

Contudo, o seaborn não aceita plotar duas séries do jeito que fizemos, então estamos vendo somente um parte dos dados. 

O que é necessário fazer ? Mudar a formatação da nossa tabela, para conseguimos plotar as duas séries no mesmo gráfico: nossa loja e a concorrência. 

```python
# Reorganizando nosso dataframe para conseguir fazer o gráfico de barras com duas séries
df_3 = pd.melt(df_2, id_vars = "Categoria", var_name= "Lojas", value_name = "Satisfação")
df_3
```

```python
# Ajustando a Coluna de Satisfação para porcentagem 
df_3["Satisfação"]=df_3["Satisfação"]*100
```

Com o dataframe reorganizado podemos plotar as duas séries, utilizando o seaborn, atraves do parâmetro **hue**. 

Ele 'divide' o gráfico de acordo com a coluna que indicamos. Neste caso, a coluna é a Loja e o gráfico ficará dividido em duas séries: Nossa Sorveteria e Concorrência.  

```python
# Criando o Objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110)   
#Plotando os dos dados  
sns.barplot(y = "Categoria", x = "Satisfação",hue = "Lojas", data = df_3, ax = ax )

# Ajustando os rótulos dos eixos 
ax.set_ylabel('Categorias') 
ax.set_xlabel('Percentual de Satisfação(%)') 

# Título do gráfico
ax.set_title("Pesquisa de Satisfação da nossa loja frente à concorrência", fontsize = 18)
```

O gráfico atual já fornece uma leitura dos dados muito mais clara. Entretanto, alguns pontos podem ser melhorados para despoluir o gráfico e destacar o que queremos na figura. 

Estes pontos:


*   Podemos remover as bordas(eixos) desncessários ao gráfico
*   A legenda automática está sobrepondo os dados
*   A escolha de cores automática não destaca os dados que estamos analisando prioritariamente - os da nossa sorveteria 
*   Podemos ordenar o gráfico da categoria de melhor desempenho a de pior desempenho - isto torna o gráfico muito mais intuitivo de se interpretar

Além dos pontos levantados, o líder da sua equipe pediu para destacar quais categorias alcançam a meta interna de pelo menos 75% de satisfação dos clientes. 



```python
# Criando a ordem em que vamos dispor nossas categorias no gráfico
a = df_3[df_3["Lojas"]=="Nossa Sorveteria "].sort_values("Satisfação",ascending = False).index
ordem = [df_3["Categoria"][i] for i in a]
ordem
```

```python
# Criando o Objeto 
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110)    
# Plotando os dados 
cores = {"Nossa Sorveteria " :"#242A8B", "Concorrência ": "#BFBEBC"}  # criando a paleta para cada umas das séries que vamos plotar no gráfico
sns.barplot(y = "Categoria", x = "Satisfação",hue = "Lojas", data = df_3, ax = ax ,order=ordem, palette= cores)

# Retirando os eixos superior e direito
sns.despine() # esse é o default dele ele retira o superior e o direito e mantém o resto

# Modificar a cor dos eixos inferior e esquerdo e dos ticks para cinza
ax.spines['bottom'].set_color('#C7BBBF')
ax.spines['left'].set_color('#C7BBBF')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

# Ajustando os rótulos dos eixos
ax.set_ylabel("")
ax.set_xlabel("Percentual de Satisfação(%)", color = "#919191", fontsize = 12)

# Título do Gráfico
ax.set_title("Pesquisa de Satisfação da nossa loja frente à concorrência", fontsize = 18, loc = "left", y = 1.1, x = -0.3)

# Ajustando a Legenda 
ax.legend(bbox_to_anchor=(-0.02,0.98), loc="lower left", ncol=2, fontsize = 12, frameon=False)

# Adicionando uma linha para enfatizar a meta de satisfação a ser batida
ax.axvline(x= 75, color = "lightgray", linestyle = "--")

# Importante quando estiver usando legenda fora do eixo 
fig.savefig('exemplo.png',  bbox_inches='tight')
```

Pontos Modificados:



*   Ao tornarmos todos os eixos, ticks e rótulos cinza destacamos as barras que são a PRINCIPAL informação do gráfico
*   Os dados da concorrência são para efeitos comparativos, mas estamos analisando os dados da **nossa loja**, então ao colocarmos as barras da concorrência em cinza, podemos comparar mas destacamos os nossos dados.
*   A legenda no início do gráfico favorece a rápida indentificação dos rótulos


Pontos que ainda podem ser melhorados:



*   Talvez colocar o eixo-x no topo favoreça a leitura dos dados
*   O título pode chamar mais atenção para o que queremos mostrar neste gráfico
*   A linha da meta não agregou muita informação neste gráfico


O que podemos fazer: **Separar as análises**


1.   Primeiro: analisar nosso desempenho frente à concorrência
2.   Segundo: analisar os pontos de performance da nossa loja







```python
# Criando o objeto 
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110) 
# Plotando nossos dados   
cores = {"Nossa Sorveteria " :"#242A8B", "Concorrência ": "#BFBEBC"}
sns.barplot(y = "Categoria", x = "Satisfação",hue = "Lojas", data = df_3, ax = ax ,order=ordem, palette= cores)

# Retirando os eixos que não vamos usar
sns.despine(bottom = True, top= False)  
""" como movemos o eixo-x para o topo, temos que "forçar" 
a retirada do eixo de inferior e não retirar o superior """
 
# Título do Gráfico
ax.set_title("Satisfação: Nossa sorveteria vs concorrência", fontsize = 16, loc = "left", y = 1.2, x = -0.4)

# Legenda do Gráfico
plt.legend(bbox_to_anchor=(-0.02,1.1), loc="lower left", ncol=2, fontsize = 12, frameon=False)

# Modificar a cor dos eixos inferior e esquerdo e dos ticks para cinza
ax.spines['top'].set_color('#C7BBBF')
ax.spines['left'].set_color('#C7BBBF')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

# Retirando os rótulos dos dois eixos
ax.set_ylabel("")
ax.set_xlabel("")

# Forçando o eixo-x para o eixo superior da nossa figura
ax.xaxis.tick_top()

# Adicionando a descrição dos dois eixos do gráfico
ax.text( -0.03 , -1.25,"Satistação(%)", fontsize = 10, color ="#919191")
ax.text( -30 , -1.25,"Categorias...", fontsize = 10, color ="#919191")
```

Pontos modificados:



*   Eixo-x no top do gráfico
*   Rótulos modificados de cada um dos eixos
*   Troca do título


Agora, podemos analisar os pontos de desempenho somente da nossa loja. **Queremos destacar para os funcionarios as categorias que se encontram abaixo da meta.** Para que assim, possamos tomar medidas para mudar este comportamento.



```python
# Criando o objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110)    

# Queremos que as categorias que estejam acima da meta estejam em cinza e as abaixo em azul
first_color = '#BFBEBE'
other_colors = '#242A8B'
cores = 5*[first_color] + 5*[other_colors]
# Plotando os dados 
sns.barplot(y = "Categoria", x = "Satisfação", data = df_3[df_3["Lojas"]=="Nossa Sorveteria "], ax = ax ,order=ordem, palette= cores)
"""" Queremos plotar somente os dados referentes à sorveteria por isso filtramos o nosso dataframe """

# Retirando os eixos
sns.despine(bottom = True, top= False)

# Título do Gráfico
ax.set_title("Pontos que precisamos melhorar em nossa loja", fontsize = 16, loc = "left", y = 1.13, x = -0.4, color = "#012136")

# Legenda do Gráfico
ax.legend(bbox_to_anchor=(-0.02,1.1), loc="lower left", ncol=2, fontsize = 12, frameon=False)

# Modificar a cor dos eixos inferior e esquerdo e dos ticks para cinza
ax.spines['top'].set_color('#C7BBBF')
ax.spines['left'].set_color('#C7BBBF')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

# Retirando os rótulos dos dois eixos
ax.set_ylabel("")
ax.set_xlabel("")

#Forçando o eixo-x para o eixo superior da nossa figura
ax.xaxis.tick_top()

# Adicionando a descrição dos eixos
ax.text( -0.03 , -1.2,"Satistação(%)", fontsize = 10, color ="#919191")
ax.text( -30 , -1.2,"Categorias...", fontsize = 10, color ="#919191")

# Adicionando a linha para enfatizar as categorias que estão abaixo da meta
ax.axvline(75, color = "lightgray", linestyle = "--")

# Adicionando a descrição da linha
ax.text( 74 , 8,"Meta\n de Satisfação", fontsize = 12, color ="#002A43", ha = "right")
```

Mantemos a mesma formatação do gráfico anterior. 

Entretanto, agora destacamos a linha que indica a meta da loja de satisfação do cliente. 

Além disso, o destaque do gráfico são as barras em azul, que representam as categorias que se encontram abaixo da meta.

Após a análise deste gráfico, desejamos contruir frases de impactos para as metas do próximo ano e pontos que podemos melhorar. 

```python
"""Criando o nosso objeto"""
fig, ax = plt.subplots(figsize=(7, 5),nrows = 1, ncols= 1,dpi=110)
# figsize - tamanho da figura , dpi - resolução da imagem, quanto maior mais pesada e mais demora a rodar

# Removendo as bordas do gráfico
for spine in plt.gca().spines.values():
    spine.set_visible(False)
# Removendo os eixos do gráfico
ax.axis('off')

# Título do gráfico
ax.text(0.02, 0.85,
         'Metas do próximo ano         ',
         fontsize=26,
         color='white',
         # Colocando um retangulo de cor envolta do texto - para dar destaque ao título
         bbox={'facecolor': '#828282', 'pad': 10, 'edgecolor': 'none'})

# Adicionando nossa frase
ax.text(0.05, 0.65,
         'Atingir, pelo menos, ',
         fontsize=19,
         color='#4A81BF')

# Adicionando nosso número que queremos destacar na apresentação - Negrito
ax.text(0.02, 0.30,
          "75%",
         fontsize=100,
         color='#4A81BF',weight='bold')

# Adicionando nossa frase
ax.text(0.05, 0.18,
         'de satisfação em todas as categorias.',
         fontsize=19,
         color='#4A81BF')
```

```python
"""Criando o nosso objeto"""
fig, ax = plt.subplots(figsize=(7, 5),nrows = 1, ncols= 1,dpi=110)
# figsize - tamanho da figura , dpi - resolução da imagem, quanto maior mais pesada e mais demora a rodar

# Removendo as bordas do gráfico
for spine in plt.gca().spines.values():
    spine.set_visible(False)
# Removendo os eixos do gráfico
ax.axis('off')

# Título do gráfico
ax.text(0.02, 0.85,
         'Investimentos         ',
         fontsize=26,
         color='white',
         # Colocando um retangulo de cor envolta do texto - para dar destaque ao título
         bbox={'facecolor': '#828282', 'pad': 10, 'edgecolor': 'none'})

# Adicionando nossa frase
ax.text(0.05, 0.65,
         'Melhorando a ',
         fontsize=19,
         color='#4A81BF')

# Adicionando nosso número que queremos destacar na apresentação - Negrito
ax.text(0.05, 0.45,
          "DIVERSIDADE",
         fontsize=50,
         color='#4A81BF',weight='bold')

# Adicionando nossa frase
ax.text(0.05, 0.35,
         'dos nossos produtos nos destacamos ',
         fontsize=19,
         color='#4A81BF')

ax.text(0.05, 0.25,
         'em relação à concorrência.',
         fontsize=19,
         color='#4A81BF')
```

Agora o líder da sua equipe pediu um gráfico que destaque as categorias a nossa loja desempenha melhor em relação à concorrencia. 

Para isso, o time resolveu mostrar o gráfico não mais em relação ao níveis absolutos de satisfação, mas agora em termos da diferença entre os niveis de satisfação das duas lojas. 

```python
# Calculando a diferença entre a satisfação dos clientes da nossa sorveteria e a concorrência

for i in df_3[df_3["Lojas"]=="Nossa Sorveteria "].index:
  df_3.loc[i,"Diferença"] = df_3.loc[i,"Satisfação"] -df_3.loc[i+10,"Satisfação"]

df_3
```

```python
# Temos que ordenar novamente agora pela diferença da satisfação
a = df_3[df_3["Lojas"]=="Nossa Sorveteria "].sort_values("Diferença",ascending = False).index
ordem = [df_3["Categoria"][i] for i in a]
ordem
```

```python
# Criando o objeto
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize=(6,6),       # (largura, altura) em polegadas 
                       dpi=110)   
 
# Queremos que as categorias que estejam com um bom desempenho - melhor que a concorrência em azul- o resto em cinza
first_color = '#242A8B'
other_colors = '#BFBEBE'
cores = 5*[first_color] + 5*[other_colors]

# Plotando os dados 
sns.barplot(y = "Categoria", x = "Diferença", data = df_3[df_3["Lojas"]=="Nossa Sorveteria "], ax = ax ,order=ordem, palette= cores)

# Retirando o eixo inferior e o direito 
sns.despine(bottom = True, top= False)

# Título do Gráfico
ax.set_title("Em que nos destacamos da concorrência?", fontsize = 16, loc = "left", y = 1.2, x = -0.4, color = "#012136")

# Legenda
ax.legend(bbox_to_anchor=(-0.02,1.1), loc="lower left", ncol=2, fontsize = 12, frameon=False)

# Modificar a cor dos eixos inferior e esquerdo e os ticks para cinza
ax.spines['top'].set_color('#C7BBBF')
ax.spines['left'].set_color('#C7BBBF')
ax.tick_params(axis='both', colors ='#919191',labelsize=10)

# Removendo os rótulos dos eixos
ax.set_ylabel("")
ax.set_xlabel("")

# Forçando o eixo-x para o eixo superior do gráfico
ax.xaxis.tick_top()

# Adicionando os textos de descrição para tornar o gráfico mais facil de se interpretar 
ax.text( -55 , -1.25,"Diferença (%) Favorável vs Concorrência", fontsize = 12, color ="#919191")
ax.text( -80 , -1.25,"Categorias...", fontsize = 12, color ="#919191")
ax.text( -55 , -1.85,"Baixo Desempenho   |", fontsize = 12, color ="#919191") # a legenda está da mesma cor que as barras equivalentes no gráfico- cinza
ax.text( -20 , -1.85,"   Ótimo Desempenho", fontsize = 12, color ="#242A8B") # a legenda está da mesma cor que as barras equivalentes no gráfico- azul
ax
```
